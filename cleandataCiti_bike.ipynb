```python
citi_bike_new = [citi_bike_dec23, citi_bike_jun24]
```

```python
citi_bike_merged = pd.concat(citi_bike_new, axis=0)

citi_bike_merged
```

```python
citi_bike_merged.isnull().sum()
```

```python
citi_bike_merged.dropna(how='all',subset=['start_station_name', 'start_station_id'], inplace=True)
```

```python
citi_bike_merged.dropna(how='all',subset=['end_station_name','end_station_id'], inplace=True)
```

```python
citi_bike_merged.dropna(how='all',subset=['end_lng', 'end_lat'], inplace=True)
```

```python
citi_bike_merged.isnull().sum()
```

```python
citi_bike_merged = citi_bike_merged.dropna(how='any')
```

```python
citi_bike_merged.isnull().sum()
```

```python
citi_bike_merged.count()
```

```python
citi_bike_merged['started_at'] = pd.to_datetime(
    citi_bike_merged['started_at'],
    errors='coerce',
    infer_datetime_format=True
)
```

```python
citi_bike_merged['ended_at'] = pd.to_datetime(
    citi_bike_merged['ended_at'],
    errors='coerce',
    infer_datetime_format=True
)
```

```python
citi_bike_clean = citi_bike_merged
```
